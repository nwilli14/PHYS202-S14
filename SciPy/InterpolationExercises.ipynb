(a) Let's turn this look-up table into a continuous function. Write a reusable function Temp(Resistance) that takes any value of Resistance and returns the interpolated temperature. Be sure to include a docstring in your function.

```python
temp = arange(10,40)
res = [3.239, 3.118, 3.004, 2.897, 2.795, 2.700, 2.610, 2.526, 2.446, 2.371, 2.300,
       2.233, 2.169, 2.110, 2.053, 2.000, 1.950, 1.902, 1.857, 1.815, 1.774, 1.736,
       1.700, 1.666, 1.634, 1.603, 1.574, 1.547, 1.521, 1.496]
```

```python
plt.plot(res, temp)
plt.show()
```

```python
from scipy.interpolate import interp1d 
def Temp(Resistance):
    """Return the Temputure for a resistance value
    for a diode
    
    Units are measured in degree [Celsius] 
    if imput units are [mega Ohms]"""
    
    interp = interp1d(res[::-1], temp[::-1])
    return interp(Resistance)

print Temp(3.004)
```
