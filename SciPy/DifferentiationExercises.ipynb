## Question 1

(a) Write two functions called twoPtForwardDiff(x,y) and twoPtCenteredDiff(x,y) that take as arguments arrays of values x and y(x) and return the derivatives of y with respect to x. Make sure that the arrays returned by the function have the same shape as the input arrays.

```python
def twoPtForwardDiff(x,y):
    """docstring"""
    dy = np.diff(y)
    dx = np.diff(x)
    dydx = np.zeros(y.shape, float)
    
    dydx[0:-1] = dy/dx
    dydx[-1] = (y[-1] - y[-2])/(x[-1] - x[-2])
    
    return dydx
```

```python
def twoPtCenteredDiff(x,y):
    """docstring"""
    dy = np.diff(y)
    dx = np.diff(x)
    dydx = np.zeros(y.shape, float)
    
    dydx[1:-1] = (y[2:] - y[:-2])/(x[2:] - x[:-2])
    dydx[0] = (y[1] - y[0])/(x[1] - x[0])
    dydx[-1] = (y[-1] - y[-2])/(x[-1] - x[-2])
    
    return dydx
```

(b) Use your functions to take the derivative of the function y=8x4+3x over the range 0 <x< 10, with steps of 0.1. How do you know they work correctly? (Hint: compare to the analytical result).

```python
x = np.arange(0,10.1,.1)
y = 8*x**4 + 3*x
dydxF = twoPtForwardDiff(x,y)
dydxC = twoPtCenteredDiff(x,y)
```

```python
dydxR = 32*x**3 + 3
plt.plot(x, dydxF, color='r', label="forward diff")
plt.plot(x, dydxC, color='b', label='center diff')
plt.plot(x, dydxR, color='g', label='real diff')
plt.legend(loc='upper left')
plt.show()
```

c) What is the accuracy of the results if you use a spacing of 0.01? (Plot the relative error)

```python
x2 = np.arange(0,10.01,.01)
y = 8*x2**4 + 3*x2
dydxR = 32*x2**3 + 3
dydxF = twoPtForwardDiff(x2,y)
dydxC = twoPtCenteredDiff(x2,y)
```

```python
plt.plot(x2, dydxF, color='r', label="forward diff")
plt.plot(x2, dydxC, color='b', label='center diff')
plt.plot(x2, dydxR, color='g', label='real diff')
plt.legend(loc='upper left')
plt.show()
```

```python
percentErrorF = 100*abs( dydxF[1:] - dydxR[1:] )/ dydxR[1:]
percentErrorC = 100*abs( dydxC[1:] - dydxR[1:] )/ dydxR[1:]

#Use a semilog y-axis to see the variation better 
plt.semilogy(x2[1:], percentErrorF, label='forward difference')
plt.semilogy(x2[1:], percentErrorC, label='center difference')
plt.ylabel('percent error with actual dydx')
plt.legend(loc="upper right")
plt.show()
```
