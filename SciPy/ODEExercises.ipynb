###### (a) See what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs t for 5 different initial angles, θ0 , spanning 0 to π. (i.e. 0, π/4 , π/2 , 3π/4 , and π ). Put them all on the same plot. What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4? What happens to the trajectory when the initial angle is 0 and π ?

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
g = 9.8 #m/s^2
ell = 0.5 #m
th0= [0, pi/4, pi/2, 3*pi/4, pi] 

def deriv(theta_current,t): # return derivatives of the array theta
    g = 9.8 #m/s^s
    ell = 0.5 #m
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta= omega
    domega= -(g/ell)*np.sin(theta)
    return np.array([dtheta, domega])

t = np.arange(0.,10.+0.01,0.01)

omega0 = 0. #starting from rest, omega = 0

theta_init0 = np.array([th0[0], omega0])
theta0 = odeint(deriv, theta_init0, t)

theta_init1 = np.array([th0[1], omega0])
theta1 = odeint(deriv, theta_init1, t)

theta_init2 = np.array([th0[2], omega0])
theta2 = odeint(deriv, theta_init2, t)

theta_init3 = np.array([th0[3], omega0])
theta3 = odeint(deriv, theta_init3, t)

theta_init4 = np.array([th0[4], omega0])
theta4 = odeint(deriv, theta_init4, t)
```

```python
plt.plot(t,theta0[:,0],'r',label='Initial Angle = 0')
plt.plot(t,theta1[:,0],'b',label='Initial Angle = $\pi$/4')
plt.plot(t,theta2[:,0],'g',label='Initial Angle = $\pi$/2')
plt.plot(t,theta3[:,0],'black',label='Initial Angle = 3$\pi$/4')
plt.plot(t,theta4[:,0],'purple',label='Initial Angle = $\pi$')
plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
plt.grid()
plt.legend(loc="upper left", bbox_to_anchor=(1,1), fontsize=15)
plt.show()
```

#### What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4?  

The period is increased.

When the th0 is 0 the pendulum does not move.

## Question 2

(a) Use Euler's method (not odeint) to solve the following first-order differential equation with given initial condition y(0) = 0.
dydx=x+2y
Find the value of y when x = 1, y(1), using two different step sizes: 0.25 then 0.02.

```python
def EulerFreefall(dx):
    xmax = 2  
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    F = x + 2*y
    
    # set initial conditions, redundant in this example
    x[0] = 0.
    y[0] = 0.

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        y[i+1] = y[i] + dx * (x[i]+2*y[i])
       
        
    return x,y
```

```python
xe,ye = EulerFreefall(.25)

from scipy.interpolate import interp1d

interp = interp1d(xe, ye)

newx = 1
newy = interp(newx)

plt.plot(xe, ye, newx, newy, 'r+', markersize=20, markeredgewidth=3)
plt.plot(xe,ye,"g-",label='y(t)')
plt.title("y'=x+2y,  dx=.25", fontsize=20)
plt.grid()
plt.show()
```

```python
xe2,ye2 = EulerFreefall(.02)

from scipy.interpolate import interp1d

interp = interp1d(xe2, ye2)

newx = 1
newy = interp(newx)

plt.plot(xe2, ye2, newx, newy, 'r+', markersize=20, markeredgewidth=3)
plt.plot(xe2,ye2,"b-",label='y(t)')
plt.title("y'=x+2y,  dx=.02", fontsize=20)
plt.grid()
plt.show()
```

(b) Plot the results from part (a) for the two different step sizes with the exact solution
y=0.25e2x−0.5x−0.25
on one plot. Include a legend. Which step size gives a more accurate result?

```python
soln = 0.25*exp(2*xe2)-0.5*xe2-0.25
plt.plot(xe,ye,"g-",label='dx=0.25')
plt.plot(xe2,ye2,"b-",label='dx=0.02')
plt.plot(xe2, soln, "r", label='exact soln')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

(c) Now solve the same problem using Euler-Cromer and the Midpoint method for the same pair of step sizes and plot the results along with the exact solution. Then also plot the difference between the approximate numerical solutions and the exact solution. Which combination of method and step-size gives the best result?

```python
def EulerCromerFreefall(dx):
    xmax = 2  
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    F = x + 2*y
    
    # set initial conditions, redundant in this example
    x[0] = 0.
    y[0] = 0.

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        y[i+1] = y[i] + dx * (x[i+1]+2*y[i+1])
       
        
    return x,y
```

```python
xe,ye = EulerCromerFreefall(.25)
xe2,ye2 = EulerCromerFreefall(.02)

soln = 0.25*exp(2*xe2)-0.5*xe2-0.25
plt.plot(xe,ye,"g-",label='dx=0.25')
plt.plot(xe2,ye2,"b-",label='dx=0.02')
plt.plot(xe2, soln, "r", label='exact soln')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
error = soln - ye2
plt.plot(xe2, error, "black", label='Euler-Cromer error of dx=0.2')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
def Midpoint(dx):
    xmax = 2  
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    F = x + 2*y
    
    # set initial conditions, redundant in this example
    x[0] = 0.
    y[0] = 0.

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        y[i+1] = y[i] + dx * (.5*(x[i+1]+x[i])+(y[i+1]+y[i]))
       
        
    return x,y
```

```python
xe,ye = Midpoint(.25)
xe2,ye2 = Midpoint(.02)

soln = 0.25*exp(2*xe2)-0.5*xe2-0.25
plt.plot(xe,ye,"g-",label='dx=0.25')
plt.plot(xe2,ye2,"b-",label='dx=0.02')
plt.plot(xe2, soln, "r", label='exact soln')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
error = soln - ye2
plt.plot(xe2, error, "black", label='midpoint error of dx=0.2')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
