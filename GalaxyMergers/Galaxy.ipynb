# Galaxy Merger Project

```python
%pylab inline
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import odeint
from scipy.integrate import ode
```

```python
#initial conditions/constants
G = 4.494e-8 #kpc^3/(SolarMass/(10^8 years)^2)
M = 10**11 #solarMasses
Rmin = 20 #kpc
stars = [12,18,24,30,36] #the number of stars in each ring 
dist = [.2,.3,.4,.5,.6] #percent difference of Rmin that each ring will be placed

#Initialize two arrays of size 120
r0 = np.zeros(120) 
theta0 = np.zeros(120)

'''This for loop places the stars on each ring 
The enumerate commmand will take the index in the list and will also take the
value of the list at that point, so si is the index, initially 0, and n will be
the value of the list at index 0, initially 12

The r0[i] determines which of the values in the array r0 should be overwritten.
the dist[si]*Rmin tells how far away the star is to be placed
the Theta's are spaced evenly around the circle'''
i = 0
for si,n in enumerate(stars):
    for x in range(n):
        r0[i] = dist[si]*Rmin
        theta0[i] = np.linspace(0,2*np.pi,n)[x]
        i += 1

#create arrays of the radial and tangential velocities
v = sqrt(G*M/r0)
vtheta = theta0 + np.pi/2

#convert from polar to cartesian 
x0 = r0*np.cos(theta0)
y0 = r0*np.sin(theta0)
vx0 = v*np.cos(vtheta)
vy0 = v*np.sin(vtheta)

#plot the stars 
plt.figure(figsize=(6,6))
plt.scatter(x0,y0)
plt.quiver(x0,y0,vx0,vy0)
plt.grid()
plt.show()
```

### Test that the galaxy orbits without a disrupter

```python
#Galaxy S initial conditions
Rx0 = -120
Ry0 = 120
dRx0 = 10
dRy0 = -5
 
initarray = np.array([Rx0, Ry0, dRx0, dRy0, x0[0], y0[0], vx0[0], vy0[0]])
 
def diff_eq_1(t, initarray):
    """This function will take an initial array of values and then return 
    the values after they have been run through a pair of differental 
    equations
    Rx0 is the x component of the inital distance of the Galaxy S from Galaxy M
    Ry0 is the y component of the inital distance of the Galaxy S from Galaxy M
    dRx0 is the x component of the initial velocity of Galaxy S
    dRx0 is the x component of the initial velocity of Galaxy S
    x0 is the x component of the initial distance of the stars from the center of Galaxy M
    y0 is the y component of the initial distance of the stars from the center of Galaxy M
    vx0 is the x component of the initial velocity of the stars
    vy0 is the y component of the initial velocity of the stars"""
    #constants
    M = 10**11 #kpc
    gamma = 4.494e-8 #kpc^3/(Solar Mass)/(10^8 years)^2
    S = 0 #no mass in the disrupter 
    
    #give local variable names to the values from the array
    x = initarray[0]
    y = initarray[1]
    dx = initarray[2]
    dy = initarray[3]
    Rx = initarray[4]
    Ry = initarray[5]
    dRx = initarray[6]
    dRy = initarray[7]
    
    #find the distanct from the center of Galaxy M to the Star
    r = np.sqrt(x**2 + y**2)
    #find the distanct from the center of Galaxy M to Galaxy S
    R = np.sqrt(Rx**2 + Ry**2)
    
    #find the distanct from the Star to Galaxy S
    rhox = Rx - x
    rhoy = Ry - y
    rho = R - r
    
    #Solve the Actual Differental equations
    dvx = -gamma * ((M/r**3)*x - (S/rho**3)*rhox+ S/(R**3)*Rx)
    dvy = -gamma * ((M/r**3)*y - (S/rho**3)*rhoy+ S/(R**3)*Ry)
    dvRx = -gamma * (M+S)/(R**3) * Rx
    dvRy = -gamma * (M+S)/(R**3) * Ry
    
    return np.array([dx, dy, dvx, dvy, dRx, dRy, dvRx, dvRy])
```

```python
#run ODE on the function diff_eq_1
r = ode(diff_eq_1).set_integrator('vode', method='bdf',max_step=5000)

#initalize time
t_start = 0.0
t_final = 20.0
delta_t = 1./2.
 
steps = np.floor((t_final - t_start)/delta_t) + 1
 
#create initial arrays
t = np.zeros((steps))
x = np.zeros((120,steps))
y = np.zeros((120,steps))
vx = np.zeros((120,steps))
vy = np.zeros((120,steps))
Rx = np.zeros((steps))
Ry = np.zeros((steps))
vRx = np.zeros((steps))
vRy = np.zeros((steps))
 
#find inital conditions
t[0] = t_start
x[:,0],y[:,0],vx[:,0],vy[:,0] = x0,y0,vx0,vy0
Rx[0],Ry[0],vRx[0],vRy[0] = Rx0,Ry0,dRx0,dRy0

#loop the data and integrate at each time step
for i,(x0i,y0i,vx0i,vy0i) in enumerate(zip(x0,y0,vx0,vy0)):
    initarray = np.array([x0i,y0i,vx0i,vy0i,Rx0,Ry0,dRx0,dRy0])
    r.set_initial_value(initarray, t_start)
    k = 1
    while r.successful() and k < steps:
        r.integrate(r.t + delta_t)
        t[k] = r.t
        x[i,k] = r.y[0]
        y[i,k] = r.y[1]
        vx[i,k] = r.y[2]
        vy[i,k] = r.y[3]
        Rx[k] = r.y[4]
        Ry[k] = r.y[5]
        vRx[k] = r.y[6]
        vRy[k] = r.y[7]
        k += 1
```

```python
#plot each indivigual star at each timestep
for i, j in enumerate(t):
    plt.figure(figsize = (5, 5))
    plt.grid()
    plt.title('time = %.1f'%j)
    plt.xlim(-20, 20)
    plt.ylim(-20, 20)
    plt.scatter(x[:, i], y[:, i])
    plt.show()
    
```

This shows that the stars continue to orbit if they are not disrupted by another Galaxy
